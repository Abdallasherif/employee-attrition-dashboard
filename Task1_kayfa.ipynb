# Preprocessing & Cleaning 🧼🧹

## Copy and Normalize column names to `snake_case`

```python
df_cleaned = df.copy()
df_cleaned.columns = (
    df_cleaned.columns
      .str.lower()
      .str.strip()
      .str.replace(" ", "_")
      .str.replace('-', '_')
)
```

## Data Quality checks:
*   missing values
*   duplicates
*   empty strings
*   column cleanup



```python
display(df_cleaned.isnull().sum())
display(df_cleaned.duplicated().sum())
display((df_cleaned.astype(str).apply(lambda x: x.str.strip()) == '').sum())
```

```python
df_cleaned.drop("employee_id", axis=1, inplace=True)
```

## Verify Category Values

```python
for col in df_cleaned.select_dtypes(include="object"):
    print(col)
    print(df_cleaned[col].unique())
```

```python
for col in df_cleaned.select_dtypes(include="object"):
    print(f"\n{'='*40}")
    print(col.upper())
    print(df_cleaned[col].value_counts())
```

## Encode Ordinal Columns

```python
ordinal_mappings = {
    "work_life_balance": {"Poor": 0, "Fair": 1, "Good": 2, "Excellent": 3},
    "company_reputation": {"Poor": 0, "Fair": 1, "Good": 2, "Excellent": 3},

    "job_satisfaction": {"Low": 0, "Medium": 1, "High": 2, "Very High": 3},
    "employee_recognition": {"Low": 0, "Medium": 1, "High": 2, "Very High": 3},

    "performance_rating": {"Low": 0, "Below Average": 1, "Average": 2, "High": 3},


    "education_level": {"High School": 0, "Associate Degree": 1, "Bachelor’s Degree": 2, "Master’s Degree": 3, "PhD": 4},

    "job_level": {"Entry": 0, "Mid": 1, "Senior": 2},
    "company_size": {"Small": 0, "Medium": 1, "Large": 2}
}
for col, mapping in ordinal_mappings.items():
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].map(mapping)
```

## Handle Binary & Target Columns (Yes/No -> 1/0)

```python
binary_cols = ['overtime', 'remote_work', 'leadership_opportunities', 'innovation_opportunities']
for col in binary_cols:
    if col in df_cleaned.columns:
        df_cleaned[col] = df_cleaned[col].map({'Yes': 1, 'No': 0})

# 6. Map Target column
if 'attrition' in df_cleaned.columns:
    df_cleaned['attrition'] = df_cleaned['attrition'].map({'Left': 1, 'Stayed': 0})
```

## Address The Anomalies & Feature Engineering (Age & Tenure groups)

```python
columns_to_drop = ['employee_id', 'company_tenure']
df_cleaned.drop(columns=[col for col in columns_to_drop if col in df_cleaned.columns], inplace=True, errors='ignore')
```

```python
df_cleaned['starting_age'] = df_cleaned['age'] - df_cleaned['years_at_company']

age_bins = [0, 29, 44, float('inf')]
age_labels = ['<30 (Young)', '30-44 (Mid-Career)', '45+ (Senior)']
df_cleaned['age_group'] = pd.cut(df_cleaned['age'], bins=age_bins, labels=age_labels)

tenure_bins = [-1, 2, 5, 10, float('inf')]
tenure_labels = ['0-2 years (New)', '2-5 years (Established)', '5-10 years (Loyal)', '10+ years (Veteran)']
df_cleaned['tenure_group'] = pd.cut(df_cleaned['years_at_company'], bins=tenure_bins, labels=tenure_labels)

print("--- Verification Check ---")
print(df_cleaned[['work_life_balance', 'job_satisfaction', 'overtime', 'attrition', 'age_group']].head())
```

```python
# We group by the column and calculate the average (mean) of attrition (0 or 1)
age_attrition = df_cleaned.groupby('age_group', observed=False)['attrition'].mean().reset_index()
tenure_attrition = df_cleaned.groupby('tenure_group', observed=False)['attrition'].mean().reset_index()

# 2. Create subplots
fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Attrition Rate by Age Group', 'Attrition Rate by Tenure Group'))

# 3. Add Age Group bar trace
fig.add_trace(
    go.Bar(x=age_attrition['age_group'], y=age_attrition['attrition'],
           marker_color='blue', name='Age Group'),
    row=1, col=1
)

# 4. Add Tenure Group bar trace
fig.add_trace(
    go.Bar(x=tenure_attrition['tenure_group'], y=tenure_attrition['attrition'],
           marker_color='navy', name='Tenure Group'),
    row=1, col=2
)

fig.update_layout(height=500, width=1000, showlegend=False, template='plotly_white')
fig.update_yaxes(title_text="Proportion of Employees Left", row=1, col=1)
fig.update_yaxes(title_text="Proportion of Employees Left", row=1, col=2)

fig.show()
```

# 5 BUSINESS INSIGHTS💡


> 📌 Insight 1 — Work Environment Drives Attrition

Employees reporting poor or fair work-life balance show significantly higher attrition rates compared to those with good or excellent balance.
👉 This suggests that improving flexibility and workload distribution could reduce employee turnover.



> 📌 Insight 2 — Low Job Satisfaction is a Major Exit Driver

Attrition is strongly concentrated among employees with low and medium job satisfaction, indicating that dissatisfaction is a primary reason employees leave.
👉 HR should focus on engagement programs and feedback systems.


> 📌 Insight 3 — Salary & Promotions Matter

Employees with lower monthly income and fewer promotions are more likely to leave the company.
👉 This indicates a need for clearer career progression and competitive compensation strategies.



> 📌 Insight 4 — The "Mid-Career" Retention Cliff

Finding: Attrition peaks significantly among employees aged 30–44.
👉 Recommendation: This group is likely hitting a "development plateau." Implement targeted mid-level leadership training and mentorship programs to improve retention.


> 📌 Insight 5 — The "New Hire" Vulnerability

Finding: Employees in their first 2 years exhibit the highest turnover risk.
👉 Recommendation: The high departure rate suggests gaps in onboarding or culture fit. Introduce 90-day pulse checks and accelerated social integration initiatives to secure long-term commitment.
